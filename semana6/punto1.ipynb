```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
q = [0., 0.2, 0.4, 0.7, 0.9, 1.]
plt.figure()
def f(u,t,q):
    return u**q
soluciones=np.zeros(len(q))
for i in range(len(q)):
    int(i)
    def f(u,t):
        return u**q[i]
    solucion= solve_ivp(f, (0,10),y0=[0]  )
    plt.plot(solucion.t,solucion.y[0], label="q="+str(q[i]))
    plt.legend()
```
